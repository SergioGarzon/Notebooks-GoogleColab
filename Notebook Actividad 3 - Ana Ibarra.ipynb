```python
# Leo el archivo de poblacion.csv
poblacion_df = pd.read_csv('poblacion.csv', encoding='latin1')

poblacion_df
```

```python
# Grafico de barras (Variable cualitativa de provincia)

# Este grafico muestra la cantidad de veces que se repite provincias dentro 

plt.figure(figsize=[30,5])
plt.hist(poblacion_df["provincia"],bins=20,color= 'lightblue',edgecolor='k')
plt.title("Distribucion de frecuencia de provincias")
```

```python
# Resolución del punto numero 2

# Diagrama de dispersión o scatterplot

# Podemos cambiar el tamaño del grafico
plt.figure(figsize=[5,5])

# Definimos el tipo de grafico y las variables
plt.scatter(poblacion_df["poblacion_varones"], poblacion_df["poblacion_mujeres"], color = 'red')

# definimos los titulos de los ejes
plt.xlabel("población de varones"); 
plt.ylabel("población de mujeres")

# titulo del grafico
plt.title("Relaciones entre poblaciones")

# Agrego lineas cuadriculada
plt.grid()
```
